```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4, 400)
seg1 = np.linspace(0, 3, 100) #entspricht Steigung 3
seg2 = np.asarray([3]*100)
seg3 = np.linspace(3, 0, 100) #TODO use -seg1
seg4 = np.asarray([0]*100)
```

```python
trapez = np.append(seg1, seg2)
trapez = np.append(trapez, seg3)
trapez = np.append(trapez, seg4)
trapez -= 1.5
```

```python
plt.plot(x, trapez)
plt.show()
```

```python
dim = len(trapez)
```

```python
# trapez[n] == y(n); x[n] == x(n)
# Keppler: 1/6*(b-a)[f(a) + 4f((a+b)/2) + f(b)]

def rechteck(b, y):
    return b*y

result = 0
b = 4/400
for i in range(dim):
    result += rechteck(b, trapez[i+1])
    print(result)
    
print(result)
```
