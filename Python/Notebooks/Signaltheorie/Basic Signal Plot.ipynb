```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
X = np.linspace(-np.pi, np.pi,  256)
C,S = np.cos(X), np.sin(X)
```

```python
plt.plot(X,C)
plt.plot(X,S)
```

```python
plt.show()
```

```python
t = np.linspace(0, 1, 500, endpoint=False)
ts = signal.square(2 * np.pi * 5 * t)
ts = signal.square(10*t*np.pi)
```

```python
plt.plot(t, ts)
plt.ylim(-2, 2)
```

```python
plt.show()
```

```python
t = np.linspace(0,  2*np.pi, 10000)
ts = signal.square(2 * np.pi *4* t)
plt.plot(t, ts + np.sin(t))
```

```python
plt.show()
```

```python
x = np.linspace(0, 8, 800) # 2 periods, each period devided into 4 segments
```

```python
seg1 = np.linspace(0, 3, 100) #entspricht Steigung 3
seg2 = np.asarray([3]*100)
seg3 = np.linspace(3, 0, 100) #TODO use -seg1
seg4 = np.asarray([0]*100)
```

```python
trapez = np.append(seg1, seg2)
trapez = np.append(trapez, seg3)
trapez = np.append(trapez, seg4)
trapez = np.append(trapez, trapez)
```

```python
plt.plot(x, trapez)
plt.show()
```
